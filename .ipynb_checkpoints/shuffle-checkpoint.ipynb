```python
import os
import random
import shutil
```

```python
dog = os.listdir("./dataset/train/dog")
print(len(dog))
i = 0
random.seed(20)
a = 12499
```

```python
while i < 250:
    fno = random.randint(0,a)
    print(fno)
    src = "./dataset/train/dog/" + dog[fno]
    shutil.move(src, "./dataset/val/dog/")
    del dog[fno]
    a = a-1
    i=i+1
print("done")
```
